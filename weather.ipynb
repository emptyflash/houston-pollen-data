```python
import requests
import pandas as pd
```

```python
pollen = pd.read_csv("pollen.csv").sort_values("Date")
pollen['Date'] = pd.to_datetime(pollen['Date'])
pollen = pollen.set_index('Date').sum(axis=1)
```

```python
api_key = "7fb27bc4d129eda8"
station = "77001.1.99999"
url = lambda date: 'http://api.wunderground.com/api/{}/history_{}/q/{}.json'.format(api_key, date.strftime("%Y%m%d"), station)

def get_weather(date):
    r = requests.get(url(date))
    json = r.json().get("history", {}).get("dailysummary", {})[0]
    json.pop('date')
    json['Date'] = date
    return json

results = [get_weather(date) for date in pollen.index]
```

```python
weather = pd.DataFrame(results)
weather.set_index('Date', inplace=True)
weather.head()
```

```python
weather.to_csv("weather.csv")
```
